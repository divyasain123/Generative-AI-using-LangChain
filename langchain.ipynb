```python
from abc import ABC,abstractmethod
```

```python
class Runnable(ABC):
  @abstractmethod
  def invoke(input_data):
    pass
```

```python
import random
class NakliLLm(Runnable):
  def __init__(self):
    print("llm created")
  def invoke(self,prompt):
    respones_list =['delhi is capital of india','mango is king of fruits','ml stands for machine learning']
    return{'response':random.choice(respones_list)}


  def predict(self,prompt):
    respones_list =['delhi is capital of india','mango is king of fruits','ml stands for machine learning']
    return{'response':random.choice(respones_list)}
```

```python
class Nakliprompt(Runnable):
  def __init__(self,template,input_variables):
    self.template=template
    self.input_variables=input_variables

  def invoke(self,input_dict):
      return self.template.format(**input_dict)
```

```python
class Runnableconnect(Runnable):

  def __init__(self,R_list):
    self.R_list=R_list

  def invoke(self,input_data):

    for runnable in self.R_list:
      runnable.invoke(input_data)

```
